
# 3. Data Cleaning and Preprocessing

Data cleaning and preprocessing are essential steps to prepare our raw data for machine learning models. In this section, we will tackle issues like inconsistent units, incorrect data types, and missing values. We'll also look at feature distributions and apply transformations to address skewness, ensuring our data is in a suitable format for training

```python
# Check if the 'mileage' column has an object data type (string).
if df['mileage'].dtype == 'object':
    # If it's an object, remove ' kmpl' and ' km/kg' units and convert the column to numeric.
    # 'errors='coerce'' will turn any values that cannot be converted into NaN.
    df['mileage'] = pd.to_numeric(df['mileage'].str.replace(' kmpl', '').str.replace(' km/kg', ''), errors='coerce')
```

```python
# Check if the 'engine' column has an object data type (string).
if df['engine'].dtype == 'object':
    # If it's an object, remove the ' CC' unit and convert the column to numeric.
    # 'errors='coerce'' will turn any values that cannot be converted into NaN.
    df['engine'] = pd.to_numeric(df['engine'].str.replace(' CC', ''), errors='coerce')
```

```python

# Check if the 'max_power' column has an object data type (string).
if df['max_power'].dtype == 'object':
    # If it's an object, remove the ' bhp' unit and convert the column to numeric.
    # 'errors='coerce'' will turn any values that cannot be converted into NaN.
    df['max_power'] = pd.to_numeric(df['max_power'].str.replace(' bhp', ''), errors='coerce')
```

```python
# Convert the 'torque' column to string type to ensure string operations can be applied.
df['torque'] = df['torque'].astype(str)
# Extract the numeric part from the 'torque' column using a regular expression and convert to numeric.
# '(\d+\.?\d*)' matches one or more digits, optionally followed by a decimal point and more digits.
# 'errors='coerce'' will turn any values that cannot be converted into NaN.
df['torque'] = pd.to_numeric(df['torque'].str.extract(r'(\d+\.?\d*)', expand=False), errors='coerce')
```

```python
# Select only the numeric columns from the DataFrame.
numeric_cols = df.select_dtypes(include=np.number).columns.tolist()

# Calculate the correlation matrix for the numeric columns.
correlation_matrix = df[numeric_cols].corr()

# Get the correlation of each numeric feature with the 'selling_price' and sort in descending order.
selling_price_correlation = correlation_matrix['selling_price'].sort_values(ascending=False)

# Print the correlation of numeric features with selling price.
print(selling_price_correlation)

# Create a heatmap to visualize the correlation matrix of numeric features.
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix of Numeric Features')
plt.show()

# Define a list of features to check for skewness.
features_to_check = ['max_power', 'engine', 'year', 'torque', 'km_driven', 'mileage']
# Calculate the skewness for the selected features.
skewness_report = df[features_to_check].skew()
# Print the skewness report.
print(skewness_report)
```

```python
# Selecting numerical features to check skewness
# Skewness shows how much the data distribution is shifted (left/right) compared to normal distribution
features_to_check = ['max_power', 'engine', 'year', 'torque', 'km_driven', 'mileage']

# Calculating skewness for the selected features
skewness_report = df[features_to_check].skew()

# Printing skewness values for each feature
print(skewness_report)
```

```python
# Defining the features that showed skewness in the earlier step
skewed_features = ['km_driven', 'engine', 'max_power', 'seats', 'year', 'torque']

# Applying log1p (log(x+1)) transformation to each skewed feature
# This helps reduce skewness and normalize the distribution
for col in skewed_features:
    df[col] = np.log1p(df[col])

# Checking skewness again after transformation
print(df[skewed_features].skew())
```

```python
# Filling missing values in 'mileage' with the column mean
df['mileage'].fillna(df['mileage'].mean(), inplace=True)
```

```python
# Filling missing values in 'engine' with the column mean
df['engine'].fillna(df['engine'].mean(), inplace=True)
```

```python
# Filling missing values in 'max_power' with the median
df['max_power'].fillna(df['max_power'].median(), inplace=True)

# Filling missing values in 'seats' with the median
df['seats'].fillna(df['seats'].median(), inplace=True)
```

In this section, we'll extract the car brand from the 'name' column as a new feature.

## 7.1 Linear Regression

We'll start with a simple yet fundamental model, Linear Regression. This model assumes a linear relationship between the features and the target variable. It's a good baseline model to understand the basic performance we can expect.

```python
# Initializing Linear Regression model
lr_model = LinearRegression()

# Applying median imputation again on train and test sets (in case of missing values)
imputer = SimpleImputer(strategy='median')
X_train_scaled_imputed = imputer.fit_transform(X_train_scaled)
X_test_scaled_imputed = imputer.transform(X_test_scaled)

# Training Linear Regression model on training data
lr_model.fit(X_train_scaled_imputed, y_train)
```

```python
# Making predictions on test data (in log scale)
y_pred_log = lr_model.predict(X_test_scaled_imputed)

# Converting log predictions back to actual selling prices
y_pred_actual = np.expm1(y_pred_log)
y_test_actual = np.expm1(y_test)

# Evaluating model performance
r2 = r2_score(y_test_actual, y_pred_actual)                     # R-squared score
mae = mean_absolute_error(y_test_actual, y_pred_actual)         # Mean Absolute Error
rmse = np.sqrt(mean_squared_error(y_test_actual, y_pred_actual))# Root Mean Squared Error

# Printing evaluation results
print("--- Linear Regression Model Performance ---")
print(f"R-squared: {r2:.4f}")
print(f"Mean Absolute Error (in REAL RUPEES): Rs. {mae:,.2f}")
print(f"RMSE (in REAL RUPEES): Rs. {rmse:,.2f}")
```

```python
# Plotting Actual vs Predicted values for Linear Regression

plt.figure(figsize=(10, 8))

# Scatter plot of actual vs predicted selling prices
sns.scatterplot(x=y_test_actual, y=y_pred_actual, alpha=0.6)

# Adding a diagonal reference line (perfect prediction line)
plt.plot([y_test_actual.min(), y_test_actual.max()],
         [y_test_actual.min(), y_test_actual.max()],
         color='red', linestyle='--', linewidth=2)

# Labels and title
plt.xlabel("Actual Selling Price")
plt.ylabel("Predicted Selling Price")
plt.title("Linear Regression: Actual vs. Predicted Prices")

# Grid and layout for clarity
plt.grid(True)
plt.tight_layout()
plt.show()
```


## 7.3 Lasso Regression

Lasso Regression is another type of linear regression that uses an L1 regularization penalty. The L1 penalty has the effect of shrinking some coefficients all the way to zero, effectively performing feature selection. This can be useful when you have many features and suspect that only a subset of them are truly relevant.

```python
from sklearn.linear_model import Lasso

lasso_instance = Lasso()
```

```python
# Importing Lasso Regression model
from sklearn.linear_model import Lasso

# Defining hyperparameter grid for alpha values to tune regularization strength
lasso_param_grid = {
    'alpha': [0.001, 0.01, 0.1, 1.0, 10.0, 100.0]
}
```

```python
lasso_random_search = RandomizedSearchCV(
    estimator=lasso_instance,        # base model (Lasso)
    param_distributions=lasso_param_grid,  # dict of hyperparameters to try
    n_iter=6,                        # sample 6 random combinations
    cv=5,                            # 5-fold cross-validation
    scoring='r2',                    # evaluate using R² score
    verbose=2,                       # print progress
    n_jobs=-1,                       # use all CPU cores
    random_state=42                  # reproducibility
)

lasso_random_search.fit(X_train_scaled, y_train)
```

```python
best_lasso_model = lasso_random_search.best_estimator_  # Get best tuned Lasso model
y_pred_lasso_tuned_log = best_lasso_model.predict(X_test_scaled)  # Predict on test set (log scale)

y_pred_lasso_tuned_actual = np.expm1(y_pred_lasso_tuned_log)  # Convert predictions back to rupees

r2_lasso_tuned = r2_score(y_test_original, y_pred_lasso_tuned_actual)  # R² score
mae_lasso_tuned = mean_absolute_error(y_test_original, y_pred_lasso_tuned_actual)  # Mean Absolute Error
rmse_lasso_tuned = np.sqrt(mean_squared_error(y_test_original, y_pred_lasso_tuned_actual))  # RMSE

print("--- Tuned Lasso Model Performance ---")
print(f"Best Alpha: {lasso_random_search.best_params_['alpha']}")  # Best alpha found
print(f"R-squared: {r2_lasso_tuned:.4f}")  # Print R²
print(f"Mean Absolute Error (in REAL RUPEES): Rs. {mae_lasso_tuned:,.2f}")  # Print MAE
print(f"RMSE (in REAL RUPEES): Rs. {rmse_lasso_tuned:,.2f}")  # Print RMSE
```

```python
lasso_cv_model = Lasso(alpha=lasso_random_search.best_params_['alpha'], random_state=42)  # Initialize Lasso with best alpha
cv_scores_lasso = cross_val_score(lasso_cv_model, X_train_scaled, y_train, cv=5, scoring='r2')  # Perform 5-fold CV

print("---- Cross-Validation R² Scores for Tuned Lasso ----")
print("CV R² scores:", cv_scores_lasso)  # Print individual CV scores
print("Mean CV R²:", cv_scores_lasso.mean())  # Print mean CV score
```


## 7.4 Random Forest Regressor

The Random Forest Regressor is an ensemble learning method that builds multiple decision trees during training and outputs the average prediction of the individual trees. This approach helps to reduce overfitting and improve the robustness and accuracy of the model compared to a single decision tree.

```python
X_train_scaled, X_test_scaled, y_train_log_split, y_test_log_split = train_test_split( X_scaled_df, y, test_size=0.2, random_state=42)  # Train-test split

y_test_original = np.expm1(y_test_log_split)  # Convert test target back to original scale
rf = RandomForestRegressor(random_state=42)  # Initialize Random Forest

param_dist = {  # Define hyperparameter search space
    'n_estimators': [100, 200, 300],
    'max_depth': [10, 15, 20, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['sqrt', 'log2'],
    'bootstrap': [True]
}

rf_random_search = RandomizedSearchCV(
    estimator=rf,  # Base model
    param_distributions=param_dist,  # Parameter grid
    n_iter=50,  # Number of random combinations to try
    cv=3,  # 3-fold cross-validation
    verbose=2,  # Print progress
    random_state=42,
    n_jobs=-1,  # Use all processors
    scoring='r2'  # Evaluation metric
)

rf_random_search.fit(X_train_scaled, y_train_log_split)  # Fit RandomizedSearchCV
```

```python
best_rf_model = rf_random_search.best_estimator_  # Get best Random Forest model
y_pred_rf_log = best_rf_model.predict(X_test_scaled)  # Predict on test data (log scale)

y_pred_rf_original = np.expm1(y_pred_rf_log)  # Convert predictions back to original scale

r2_rf_real = r2_score(y_test_original, y_pred_rf_original)  # R² score on original scale
mae_rf_real = mean_absolute_error(y_test_original, y_pred_rf_original)  # MAE on original scale
rmse_rf_real = np.sqrt(mean_squared_error(y_test_original, y_pred_rf_original))  # RMSE on original scale

print(" Best Parameters Found:")
print(rf_random_search.best_params_)  # Print best hyperparameters
print("\n Random Forest Regressor Performance (on Original Scale):")
print(f"R² Score  : {r2_rf_real:.4f}")  # Print R² score
print(f"MAE       : Rs. {mae_rf_real:,.2f}")  # Print MAE
print(f"RMSE      : Rs. {rmse_rf_real:,.2f}")  # Print RMSE

plt.figure(figsize=(10, 8))  # Set figure size
sns.scatterplot(x=y_test_log_split, y=y_pred_rf_log)  # Scatter plot (log scale)
plt.plot([y_test_log_split.min(), y_test_log_split.max()],
         [y_test_log_split.min(), y_test_log_split.max()], color='red', linestyle='--', linewidth=2)  # Perfect fit line
plt.xlabel("Actual Selling Price (Log Scale)")  # X-axis label
plt.ylabel("Predicted Selling Price (Log Scale)")  # Y-axis label
plt.title("Random Forest: Actual vs. Predicted Prices (Log Scale)")  # Title
plt.show()  # Show plot
```

```python
y_train_pred_rf_log = best_rf_model.predict(X_train_scaled)  # Predict on training set (log scale)

y_train_original = np.expm1(y_train_log_split)  # Convert actual training values back to original scale
y_train_pred_rf_original = np.expm1(y_train_pred_rf_log)  # Convert predicted training values back to original scale

r2_rf_train_real = r2_score(y_train_original, y_train_pred_rf_original)  # R² score on training set
mae_rf_train_real = mean_absolute_error(y_train_original, y_train_pred_rf_original)  # MAE on training set

print("--- Tuned Random Forest Model Performance on Training Set ---")
print(f"  R-squared: {r2_rf_train_real:.4f}")  # Print R² score for training set
print(f"  Mean Absolute Error (in REAL RUPEES): Rs. {mae_rf_train_real:,.2f}")  # Print MAE for training set

print("\n--- Tuned Random Forest Model Performance on Test Set ---")
print(f"  R-squared: {r2_rf_real:.4f}")  # Print R² score for test set
print(f"  Mean Absolute Error (in REAL RUPEES): Rs. {mae_rf_real:,.2f}")  # Print MAE for test set
```

```python
rf = RandomForestRegressor(**rf_random_search.best_params_, random_state=42)  # Initialize RF with best params
cv_scores = cross_val_score(rf, X_train_scaled, y_train_log_split, cv=5, scoring='r2')  # 5-fold CV on training set

print("Cross-Validation R² scores:", cv_scores)  # Print individual CV scores
print("Mean CV R²:", cv_scores.mean())  # Print mean CV score
```

## 7.5 LightGBM Regressor

LightGBM (Light Gradient Boosting Machine) is a gradient boosting framework that uses tree-based learning algorithms. It is designed to be distributed and efficient, with faster training speed and lower memory consumption compared to other gradient boosting frameworks. It is known for its high accuracy and ability to handle large datasets.

```python
X_train_scaled, X_test_scaled, y_train, y_test = train_test_split(
    X_scaled_df,  # Features after scaling
    y,  # Target variable (log-transformed selling price)
    test_size=0.2,  # 20% of data for testing
    random_state=42  # Ensure reproducibility
)
```

```python
lgbm_model = LGBMRegressor(random_state=42)  # Initialize LightGBM Regressor

# Clean column names (replace special characters with underscores) for training set
X_train_scaled.columns = pd.Series(X_train_scaled.columns).apply(
    lambda x: "".join(c if c.isalnum() else "_" for c in str(x))
)

# Clean column names (replace special characters with underscores) for test set
X_test_scaled.columns = pd.Series(X_test_scaled.columns).apply(
    lambda x: "".join(c if c.isalnum() else "_" for c in str(x))
)

# Handle duplicate column names in training set by appending index numbers
cols = pd.Series(X_train_scaled.columns)
for dup in cols[cols.duplicated()].unique():
    cols[cols[cols == dup].index.values.tolist()] = [
        dup + '_' + str(i) if i != 0 else dup for i in range(sum(cols == dup))
    ]
X_train_scaled.columns = cols

# Handle duplicate column names in test set by appending index numbers
cols = pd.Series(X_test_scaled.columns)
for dup in cols[cols.duplicated()].unique():
    cols[cols[cols == dup].index.values.tolist()] = [
        dup + '_' + str(i) if i != 0 else dup for i in range(sum(cols == dup))
    ]
X_test_scaled.columns = cols

lgbm_model.fit(X_train_scaled, y_train)  # Train the LightGBM model
```

## 7.6 Gradient Boosting Regressor

Gradient Boosting is a powerful ensemble technique that builds a series of decision trees sequentially. Each new tree attempts to correct the errors made by the previous ones. It's known for its high accuracy and flexibility, although it can be sensitive to hyperparameters.

```python
# Impute missing values in training and test sets using median strategy
imputer = SimpleImputer(strategy='median')
X_train_scaled = imputer.fit_transform(X_train_scaled)
X_test_scaled = imputer.transform(X_test_scaled)

# Initialize Gradient Boosting Regressor with a fixed random state
gbr_model = GradientBoostingRegressor(random_state=42)

# Define hyperparameter search space for Gradient Boosting
gbr_param_grid = {
    'n_estimators': [100, 150, 200],
    'learning_rate': [0.01, 0.05, 0.1],
    'max_depth': [3, 5, 7],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'subsample': [0.6, 0.8, 1.0],
    'max_features': ['sqrt', 'log2', None]
}

# Setup randomized hyperparameter search with 3-fold CV for Gradient Boosting
random_search_gbr = RandomizedSearchCV(
    estimator=gbr_model,
    param_distributions=gbr_param_grid,
    n_iter=50,
    cv=3,
    scoring='r2',
    verbose=2,
    n_jobs=-1,
    random_state=42
)

# Fit the randomized search on the training data
random_search_gbr.fit(X_train_scaled, y_train)
```

```python
# Get the best hyperparameters from the randomized search for Gradient Boosting
best_gbr_params = random_search_gbr.best_params_

# Print the best parameters found
print("Best GBR Parameters:")
print(best_gbr_params)

# Initialize a new Gradient Boosting model with the tuned parameters
gbr_model_tuned = GradientBoostingRegressor(**best_gbr_params, random_state=42)

# Fit the tuned Gradient Boosting model on the training data
gbr_model_tuned.fit(X_train_scaled, y_train)
```

```python
# Create a pipeline with median imputer followed by the tuned Gradient Boosting model
gbr_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='median')),
    ('gbr', gbr_model_tuned)
])

# Perform 5-fold cross-validation using R² as the scoring metric
cv_scores = cross_val_score(
    gbr_pipeline,
    X_train_scaled,
    y_train,
    cv=5,
    scoring='r2',
    n_jobs=-1
)

# Print cross-validation R² scores
print("\nCross-Validation R² scores:", cv_scores)

# Print mean of cross-validation R² scores
print("Mean CV R²:", np.mean(cv_scores))

# Print standard deviation of cross-validation R² scores
print("Standard Deviation of CV R²:", np.std(cv_scores))
```

```python
# Predict on test set using tuned Gradient Boosting model (log scale)
y_pred_test_log = gbr_model_tuned.predict(X_test_scaled)

# Convert test predictions back to original scale
y_pred_test = np.expm1(y_pred_test_log)

# Predict on training set using tuned Gradient Boosting model (log scale)
y_pred_train_log = gbr_model_tuned.predict(X_train_scaled)

# Convert training predictions back to original scale
y_pred_train = np.expm1(y_pred_train_log)

# Calculate R² and MAE for test set
r2_test = r2_score(y_test_original, y_pred_test)
mae_test = mean_absolute_error(y_test_original, y_pred_test)

# Calculate R² and MAE for training set
r2_train = r2_score(y_train_original, y_pred_train)
mae_train = mean_absolute_error(y_train_original, y_pred_train)

# Print performance metrics for training set
print("\n--- Tuned GBR Performance on Training Set ---")
print(f"  R-squared: {r2_train:.4f}")
print(f"  MAE: Rs. {mae_train:,.2f}")

# Print performance metrics for test set
print("\n--- Tuned GBR Performance on Test Set ---")
print(f"  R-squared: {r2_test:.4f}")
print(f"  MAE: Rs. {mae_test:,.2f}")
```

Elastic Net Regression is a linear regression model that combines the penalties of both Lasso (L1 regularization) and Ridge (L2 regularization).
